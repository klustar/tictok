```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('D:/termproject/countries of the world.csv', sep=',')
df2=pd.read_csv('D:/termproject/Military Expenditure.csv', sep=',')
```

```python
row_range=range(0,227)
column_range=list(range(0,19))
df=df.loc[row_range]
df=df.iloc[:,column_range]

df=df.drop(['Population'],axis=1)
df=df.drop(['Area (sq. mi.)'],axis=1)
df=df.drop(['Coastline (coast/area ratio)'],axis=1)
df=df.drop(['Net migration'],axis=1)
df=df.drop(['Infant mortality (per 1000 births)'],axis=1)
df=df.drop(['Arable (%)'],axis=1)
df=df.drop(['Crops (%)'],axis=1)
df=df.drop(['Other (%)'],axis=1)
```

```python
df
```

```python
scatter0=plt.figure()
axes0=scatter0.add_subplot(1,1,1)

axes0.scatter(df['Literacy (%)'],df['Phones (per 1000)'],color='b',s=5,marker='o')
axes0.set_title('Literacy rate and cell phone usage')
axes0.set_xlabel('Literacy')
axes0.set_ylabel('Phone')

scatter1=plt.figure()
axes1=scatter1.add_subplot(1,1,1)

axes1.scatter(df['Industry'],df['Phones (per 1000)'],color='g',s=5,marker='o')
axes1.set_title('Industry and cell phone usage')
axes1.set_xlabel('Industry')
axes1.set_ylabel('Phone')

scatter2=plt.figure()
axes2=scatter2.add_subplot(1,1,1)

axes2.scatter(df['Service'],df['Phones (per 1000)'],color='r',s=5,marker='o')
axes2.set_title('Service and cell phone usage')
axes2.set_xlabel('Service')
axes2.set_ylabel('Phone')

scatter3=plt.figure()
axes3=scatter3.add_subplot(1,1,1)

axes3.scatter(df['Agriculture'],df['Phones (per 1000)'],color='y',s=5,marker='o')
axes3.set_title('Agriculture and cell phone usage')
axes3.set_xlabel('Agriculture')
axes3.set_ylabel('Phone')
```

```python
boxplot=plt.figure()
box=boxplot.add_subplot(1,1,1)

box.boxplot(
    [df2['2000'],df2['2001'],df2['2002'],df2['2003'],df2['2004'],
    df2['2005'],df2['2006'],df2['2007'],df2['2008'],df2['2009'],
    df2['2010'],df2['2011'],df2['2012'],df2['2013'],df2['2014'],
    df2['2015'],df2['2016']],
    labels=['2000','2001','2002','2003','2004',
           '2005','2006','2007','2008','2009',
           '2010','2011','2012','2013','2014',
           '2015','2016'])
box.set_title('2000~2016 Military Expenditure')
box.set_xlabel('Year')
box.set_ylabel('USD')
```
